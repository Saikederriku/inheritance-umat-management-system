This notebook demonstrates inheritance in Python using a university management system.

## 1. Parent Class: Person

The base class that holds common attributes (`name`, `age`) and a method (`display_info`) for all people on campus.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")
````

## 2. Child Class: Student

Inherits from `Person`. Adds `student_id` and an `enroll_course()` method.
Overrides `display_info()` to include the student ID while still calling the parent's version via `super()`.

````python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
    
    def enroll_course(self, course):
        print(f"{self.name} enrolled in {course}")
    
    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")
````

## 3. Child Class: Lecturer

Inherits from `Person`. Adds `employee_id` and a `teach_course()` method.
Overrides `display_info()` to include the employee ID while still calling the parent's version via `super()`.

````python
class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id
    
    def teach_course(self, course):
        print(f"{self.name} is teaching {course}")
    
    def display_info(self):
        super().display_info()
        print(f"Employee ID: {self.employee_id}")
````

## 4. Demonstration

Create one object of each child class and demonstrate:
- Inherited `display_info()` behavior (with override)
- Unique methods (`enroll_course`, `teach_course`)

````python
# Create objects
student = Student("KENNETH DAVID KORLEY KISSEIH", 18, "FOE.41.006.097.25")
lecturer = Lecturer("Dr. Cobbinah Matthew", 45, "I DON'T KNOW")

print("--- Student ---")
student.display_info()
student.enroll_course("Object Oriented Programming")
student.display_info()

print("\n--- Lecturer ---")
lecturer.display_info()
lecturer.teach_course("Object Oriented Programming")
````
